# 程式實例ch5_5.py:請輸入數字分數，程式將回應A、B、C、D或F等級。

```python
# ch5_5.py
print("計算最終成績")
score = input("請輸入分數:")
sc = int(score)
if (sc >= 90):
    print(" A")
elif (sc >=80):
    print(" B")
elif (sc >=70):
    print(" C")
elif (sc >=60):
    print(" D")
else:
    print(" F")
```

# 練習5-1:出生年year先減去1900，再計算rem=year%12。
# rem==0是屬鼠，rem==1是屬牛，...rem==11是屬豬。
# 請試寫一程式，可輸入西元年，輸出生肖。

```python
# 練習5-1
print("計算出生西元年之生肖")
year = input("請輸入出生西元年:")
yr = int(year)
yr = yr-1900
rem = yr%12
if (rem==0):
    print("屬鼠")
elif (rem==1):
    print("屬牛")
elif (rem==2):
    print("屬虎")
elif (rem==3):
    print("屬兔")
elif (rem==4):
    print("屬龍")
elif (rem==5):
    print("屬蛇")
elif (rem==6):
    print("屬馬")
elif (rem==7):
    print("屬羊")
elif (rem==8):
    print("屬猴")
elif (rem==9):
    print("屬雞")
elif (rem==10):
    print("屬狗")
else:
    print("屬豬")
```

# 程式實例ch5_6.py:有一地區的票價收費標準是100元。
### ·但是如果小於等於6歲或大於等於80歲，收費是打2折。
### ·但是如果是7-12歲或60-79歲，收費是打5折。請輸入歲數，程式會計算票價。

```python
# ch5_6.py
print("計算票價")
age = input("請輸入年齡:")
ag = int(age)
if ag <= 6 or ag >= 80:
    print("您的票價為: 20 元")
elif (7 <= ag <= 12) or (60 <= ag <= 79):
    print("您的票價為: 50 元")
else:
    print("您的票價為: 100 元")
```
